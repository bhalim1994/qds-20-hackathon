# Top 100 most dangerous parking location

```python
dan_locations = parking_violation['location2'].value_counts()[:100].index.tolist()
```

```python
from geopy.geocoders import Nominatim
li = []
geolocator = Nominatim(user_agent="specify_your_app_name_here")
for loc in dan_locations:
    try:
        location = geolocator.geocode(loc + " Toronto, Ontario, Canada")
        print(location.latitude, location.longitude)
        li.append([location.latitude, location.longitude])
    except:
        print(loc + " Toronto, Ontario, Canada" + " doesn't return geo-location")
```

```python
dang_parking_geoloc = li
```

```python
with open('data/parking-stall-data.json') as f:
  data = json.load(f)
```

```python
from math import radians, cos, sin, asin, sqrt

def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371 # Radius of earth in kilometers. Use 3956 for miles
    return c * r
```

# Put danger property for each parking stall

```python
for park in data['carparks']:
    min_dis = 99999999999.0
    for dan in dang_parking_geoloc:
            distance = haversine(float(park['lat']), float(park['lng']), dan[0], dan[1])
            if distance < min_dis:
                min_dis = distance
    if min_dis < 0.2:
        park['danger'] = 'true'
    else:
        park['danger'] = 'false'
```
