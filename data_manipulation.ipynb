```python
import pandas as pd
import numpy as np
import json 
import collections
```

# Importing the crime data of toronto

```python
with open("data/crime-toronto.json", "r") as read_file:
    data = json.load(read_file)
```

# Making a dictionary of number of crime for each neighborhood

```python
features = data["features"]
dict_crime_by_hood = dict();
for crime in features:
    hood_id = crime["attributes"]["Hood_ID"]
    if crime["attributes"]["Hood_ID"] not in dict_crime_by_hood.keys():
        dict_crime_by_hood[hood_id] = 0
#         print("present")
#     else:
#         print("not present")
for crime in features:
    hood_id = crime["attributes"]["Hood_ID"]
    dict_crime_by_hood[hood_id] += 1
```

```python
print(dict_crime_by_hood)
```

# Importing the neighborhood data of toronto

```python
with open("data/torontohoods.geojson", "r") as read_file:
    hood_data = json.load(read_file)
```

# Making a dictionary of surface area for each neighborhood

```python
hood_area_dict = dict();
for hood in hood_data["features"]:
    code = hood["properties"]["AREA_SHORT_CODE"]
    area = hood["properties"]["Shape__Area"]
    hood_area_dict[code] = area
```

```python
print(hood_area_dict)
```

# Calculating the crime rate (by unit area) of each neighborhood

```python
hood_rate_dict = dict()
for key, value in dict_crime_by_hood.items():
    hood_rate_dict[key] = round(value / hood_area_dict[key] * 1000000, 3)
```

```python
print(hood_rate_dict)
```

```python
with open('data/crime_rate_toronto.json', 'w') as fp:
    json.dump(hood_rate_dict, fp)
```
